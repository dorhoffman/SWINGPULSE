```python
import os
import glob
import pandas as pd

PROJECT_PATH = "/content/drive/MyDrive/SWINGPULSE"

SOURCE_FOLDER = (
    f"{PROJECT_PATH}/data/raw/archive (3)"
)

CLEAN_FOLDER = (
    f"{PROJECT_PATH}/data/clean"
)

CLEAN_PATH = (
    f"{CLEAN_FOLDER}/SWINGPULSE_CLEAN_DATASET.csv"
)

os.makedirs(CLEAN_FOLDER, exist_ok=True)

csv_files = glob.glob(
    f"{SOURCE_FOLDER}/**/*.csv",
    recursive=True
)

print(f"Found {len(csv_files)} CSV files")
print("Output:", CLEAN_PATH)
```

```python
# מחיקה של קובץ קודם, אם קיים
if os.path.exists(CLEAN_PATH):
    os.remove(CLEAN_PATH)

first_write = True

total_before = 0
total_after = 0

removed_missing = 0
removed_duplicates = 0
removed_invalid_prices = 0
removed_invalid_ohlc = 0

successful_files = 0
failed_files = []

for index, file_path in enumerate(csv_files, start=1):

    try:
        df = pd.read_csv(
            file_path,
            low_memory=False
        )

        total_before += len(df)

        # ניקוי שמות עמודות
        df.columns = [
            str(column).strip()
            for column in df.columns
        ]

        df = df.rename(
            columns={
                "Adj Close": "Adj_Close",
                "Stock Splits": "Stock_Splits"
            }
        )

        # הוספת סימול לפי שם הקובץ
        ticker = os.path.splitext(
            os.path.basename(file_path)
        )[0]

        df["Symbol"] = ticker

        required_columns = [
            "Date",
            "Symbol",
            "Open",
            "High",
            "Low",
            "Close",
            "Volume"
        ]

        missing_columns = [
            column
            for column in required_columns
            if column not in df.columns
        ]

        if missing_columns:
            raise ValueError(
                f"Missing columns: {missing_columns}"
            )

        optional_columns = [
            "Adj_Close",
            "Dividends",
            "Stock_Splits"
        ]

        selected_columns = required_columns + [
            column
            for column in optional_columns
            if column in df.columns
        ]

        df = df[selected_columns].copy()

        # המרת תאריך
        df["Date"] = pd.to_datetime(
            df["Date"],
            errors="coerce",
            utc=True
        )

        # הסרת אזור זמן כדי לשמור CSV פשוט
        df["Date"] = df["Date"].dt.tz_localize(None)

        # המרת עמודות מספריות
        numeric_columns = [
            "Open",
            "High",
            "Low",
            "Close",
            "Volume"
        ]

        for column in numeric_columns:
            df[column] = pd.to_numeric(
                df[column],
                errors="coerce"
            )

        # הסרת ערכים חסרים
        before = len(df)

        df = df.dropna(
            subset=required_columns
        )

        removed_missing += before - len(df)

        # הסרת מחירים לא תקינים
        before = len(df)

        df = df[
            (df["Open"] > 0) &
            (df["High"] > 0) &
            (df["Low"] > 0) &
            (df["Close"] > 0) &
            (df["Volume"] >= 0)
        ]

        removed_invalid_prices += before - len(df)

        # בדיקת OHLC
        before = len(df)

        df = df[
            (df["High"] >= df["Low"]) &
            (df["Open"] >= df["Low"]) &
            (df["Open"] <= df["High"]) &
            (df["Close"] >= df["Low"]) &
            (df["Close"] <= df["High"])
        ]

        removed_invalid_ohlc += before - len(df)

        # הסרת כפילויות
        before = len(df)

        df = df.drop_duplicates(
            subset=["Symbol", "Date"]
        )

        removed_duplicates += before - len(df)

        # מיון
        df = df.sort_values(
            ["Symbol", "Date"]
        )

        # כתיבה ישירה ל-Drive
        df.to_csv(
            CLEAN_PATH,
            mode="w" if first_write else "a",
            header=first_write,
            index=False
        )

        first_write = False
        successful_files += 1
        total_after += len(df)

        if index % 25 == 0:
            print(
                f"Processed {index}/{len(csv_files)} files | "
                f"Clean rows: {total_after:,}"
            )

        del df

    except Exception as error:
        failed_files.append(
            {
                "file": os.path.basename(file_path),
                "error": str(error)
            }
        )

print("\nCleaning and merge completed")
```

```python
print("=" * 60)
print("SWINGPULSE DATA CLEANING SUMMARY")
print("=" * 60)

print(f"Files found:             {len(csv_files):,}")
print(f"Successful files:        {successful_files:,}")
print(f"Failed files:            {len(failed_files):,}")

print(f"\nRows before cleaning:    {total_before:,}")
print(f"Rows after cleaning:     {total_after:,}")
print(f"Rows removed:            {total_before - total_after:,}")

print("\nRemoval reasons:")
print(f"Missing values:          {removed_missing:,}")
print(f"Invalid prices:          {removed_invalid_prices:,}")
print(f"Invalid OHLC ranges:     {removed_invalid_ohlc:,}")
print(f"Duplicates:              {removed_duplicates:,}")

print("\nSaved to:")
print(CLEAN_PATH)

if failed_files:
    display(pd.DataFrame(failed_files).head(20))
```

```python
total_rows = 0
symbols = set()
min_date = None
max_date = None

for chunk in pd.read_csv(
    CLEAN_PATH,
    chunksize=100_000,
    parse_dates=["Date"],
    low_memory=False
):
    total_rows += len(chunk)

    symbols.update(
        chunk["Symbol"]
        .dropna()
        .unique()
    )

    current_min = chunk["Date"].min()
    current_max = chunk["Date"].max()

    if min_date is None or current_min < min_date:
        min_date = current_min

    if max_date is None or current_max > max_date:
        max_date = current_max

print("=" * 60)
print("SWINGPULSE DATASET OVERVIEW")
print("=" * 60)

print(f"Total rows:   {total_rows:,}")
print(f"Total stocks: {len(symbols):,}")
print(f"Start date:   {min_date}")
print(f"End date:     {max_date}")
```

```python
numeric_columns = [
    "Open",
    "High",
    "Low",
    "Close",
    "Volume"
]

count_values = pd.Series(0, index=numeric_columns, dtype="int64")
sum_values = pd.Series(0.0, index=numeric_columns)
sum_squared = pd.Series(0.0, index=numeric_columns)

min_values = pd.Series(np.inf, index=numeric_columns)
max_values = pd.Series(-np.inf, index=numeric_columns)

for chunk in pd.read_csv(
    CLEAN_PATH,
    usecols=numeric_columns,
    chunksize=100_000,
    low_memory=False
):
    for column in numeric_columns:
        values = pd.to_numeric(
            chunk[column],
            errors="coerce"
        ).dropna()

        count_values[column] += len(values)
        sum_values[column] += values.sum()
        sum_squared[column] += (values ** 2).sum()

        if not values.empty:
            min_values[column] = min(
                min_values[column],
                values.min()
            )

            max_values[column] = max(
                max_values[column],
                values.max()
            )

mean_values = sum_values / count_values

std_values = np.sqrt(
    (sum_squared / count_values)
    - (mean_values ** 2)
)

descriptive_summary = pd.DataFrame({
    "count": count_values,
    "mean": mean_values,
    "std": std_values,
    "min": min_values,
    "max": max_values
})

display(descriptive_summary)
```

```python
stock_counts = pd.Series(dtype="int64")

for chunk in pd.read_csv(
    CLEAN_PATH,
    usecols=["Symbol"],
    chunksize=100_000
):
    chunk_counts = chunk["Symbol"].value_counts()

    stock_counts = stock_counts.add(
        chunk_counts,
        fill_value=0
    )

stock_counts = (
    stock_counts
    .astype(int)
    .sort_values(ascending=False)
)

display(stock_counts.head(20))
```

```python
plt.figure(figsize=(14, 6))

stock_counts.head(20).plot(
    kind="bar"
)

plt.title("Top 20 Stocks by Number of Records")
plt.xlabel("Stock Symbol")
plt.ylabel("Number of Trading Records")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
year_counts = pd.Series(dtype="int64")

for chunk in pd.read_csv(
    CLEAN_PATH,
    usecols=["Date"],
    chunksize=100_000,
    parse_dates=["Date"]
):
    years = chunk["Date"].dt.year.value_counts()

    year_counts = year_counts.add(
        years,
        fill_value=0
    )

year_counts = (
    year_counts
    .astype(int)
    .sort_index()
)

display(year_counts)
```

```python
plt.figure(figsize=(14, 6))

year_counts.plot(
    kind="bar"
)

plt.title("Number of Records per Year")
plt.xlabel("Year")
plt.ylabel("Number of Records")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sample_parts = []

sample_fraction = 0.02
random_state = 42

for chunk in pd.read_csv(
    CLEAN_PATH,
    chunksize=100_000,
    low_memory=False
):
    sampled_chunk = chunk.sample(
        frac=sample_fraction,
        random_state=random_state
    )

    sample_parts.append(sampled_chunk)

eda_sample = pd.concat(
    sample_parts,
    ignore_index=True
)

print("Sample shape:", eda_sample.shape)

eda_sample["Date"] = pd.to_datetime(
    eda_sample["Date"],
    errors="coerce"
)
```

```python
close_upper_limit = eda_sample["Close"].quantile(0.99)

close_plot_data = eda_sample[
    eda_sample["Close"] <= close_upper_limit
]

plt.figure(figsize=(12, 6))

plt.hist(
    close_plot_data["Close"],
    bins=100
)

plt.title("Distribution of Closing Prices – Up to 99th Percentile")
plt.xlabel("Closing Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

print("99th percentile:", close_upper_limit)
```

```python
plt.figure(figsize=(10, 5))

plt.boxplot(
    close_plot_data["Close"],
    vert=False
)

plt.title("Boxplot of Closing Prices – Up to 99th Percentile")
plt.xlabel("Closing Price")
plt.tight_layout()
plt.show()
```

```python
volume_data = eda_sample[
    eda_sample["Volume"] > 0
].copy()

volume_data["log_volume"] = np.log10(
    volume_data["Volume"]
)

plt.figure(figsize=(12, 6))

plt.hist(
    volume_data["log_volume"],
    bins=100
)

plt.title("Distribution of Trading Volume – Log Scale")
plt.xlabel("Log10 Volume")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
scatter_sample = eda_sample.sample(
    n=min(20_000, len(eda_sample)),
    random_state=42
)

scatter_sample = scatter_sample[
    (scatter_sample["Close"] > 0) &
    (scatter_sample["Volume"] > 0)
]

plt.figure(figsize=(10, 6))

plt.scatter(
    np.log10(scatter_sample["Close"]),
    np.log10(scatter_sample["Volume"]),
    alpha=0.3
)

plt.title("Closing Price vs Trading Volume")
plt.xlabel("Log10 Closing Price")
plt.ylabel("Log10 Volume")
plt.tight_layout()
plt.show()
```

```python
top_symbols = stock_counts.head(10).index.tolist()

top_stock_parts = []

for chunk in pd.read_csv(
    CLEAN_PATH,
    chunksize=100_000,
    parse_dates=["Date"],
    low_memory=False
):
    filtered = chunk[
        chunk["Symbol"].isin(top_symbols)
    ]

    if not filtered.empty:
        top_stock_parts.append(filtered)

top_stocks_data = pd.concat(
    top_stock_parts,
    ignore_index=True
)

top_stocks_data = top_stocks_data.sort_values(
    ["Symbol", "Date"]
)

top_stocks_data["Daily_Return"] = (
    top_stocks_data
    .groupby("Symbol")["Close"]
    .pct_change()
)
```

```python
returns = top_stocks_data[
    "Daily_Return"
].dropna()

returns_plot = returns[
    returns.between(
        returns.quantile(0.01),
        returns.quantile(0.99)
    )
]

plt.figure(figsize=(12, 6))

plt.hist(
    returns_plot,
    bins=100
)

plt.title("Daily Return Distribution – Top 10 Stocks")
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
